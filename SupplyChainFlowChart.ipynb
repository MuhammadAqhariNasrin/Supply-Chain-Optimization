```python
import networkx as nx
import matplotlib.pyplot as plt

def create_supply_chain_flowchart(nodes, edges, node_positions=None):
    """
    Create a supply chain flowchart using networkx and matplotlib.

    Parameters:
        nodes (list): A list of node labels representing entities in the supply chain.
        edges (list of tuples): A list of edge tuples representing connections between nodes.
        node_positions (dict, optional): A dictionary specifying the positions of nodes for layout.

    Returns:
        None (displays the flowchart).
    """
    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes and edges to the graph
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)

    # Create a figure and draw the graph
    plt.figure(figsize=(8, 6))

    if node_positions is None:
        # If node positions are not provided, use the default layout
        pos = nx.spring_layout(G)
    else:
        # Use the provided node positions for layout
        pos = node_positions

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_size=1000,
        node_color="skyblue",
        font_size=12,
        font_color="black",
        font_weight="bold",
        arrows=True,
    )
    
    plt.title("Supply Chain Flowchart")
    plt.axis("off")
    plt.show()



    
```

```python
# Example usage:
nodes = ["Supplier", "Manufacturer", "Distributor", "Retailer", "Customer"]
edges = [("Supplier", "Manufacturer"), ("Manufacturer", "Distributor"),
         ("Distributor", "Retailer"), ("Retailer", "Customer")]

create_supply_chain_flowchart(nodes, edges)

    
```
